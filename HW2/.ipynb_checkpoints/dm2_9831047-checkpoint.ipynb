```python
import pandas as pd

df = pd.read_csv("housePrice.csv")
df
```

```python
# note to author: since we are using quartiles, the correct categories are under/over median and not mean!
def categorize_price(price):
    if price <= quartiles[0.25]:
        return 'cheap'
    elif price <= quartiles[0.5]:
        return 'underMedian'
    elif price <= quartiles[0.75]:
        return 'upperMedian'
    else:
        return 'expensive'

categorical_price_df = df.copy()
# add priceLevel column based on quartiles of price:
quartiles = categorical_price_df['Price'].quantile([0, 0.25, 0.5, 0.75, 1.0])
categorical_price_df['priceLevel'] = categorical_price_df['Price'].apply(categorize_price)
categorical_price_df

```

```python
# categorical to numerical:
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
numeric_df = categorical_price_df.copy()
numeric_df['priceLevel'] = le.fit_transform(numeric_df['priceLevel'])
numeric_df['Address'] = le.fit_transform(numeric_df['Address'])
numeric_df['Elevator'] = le.fit_transform(numeric_df['Elevator'])
numeric_df['Warehouse'] = le.fit_transform(numeric_df['Warehouse'])
numeric_df['Parking'] = le.fit_transform(numeric_df['Parking'])
numeric_df
```

```python
# normalize data:
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_df = numeric_df.copy()
normalized_df[["Area","Room","Parking","Warehouse","Elevator","Address","Price"]] = scaler.fit_transform(normalized_df[["Area","Room","Parking","Warehouse","Elevator","Address","Price"]])

normalized_df
```

```python
# split data:
from sklearn.model_selection import train_test_split
train, test = train_test_split(normalized_df, test_size=0.2, random_state=42)
normalized_df.size , train.size, test.size
```

```python
# linear regression:
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

lin_reg = LinearRegression()
lin_reg.fit(train[['Area','Room','Parking','Warehouse','Elevator','Address']], train['Price'])
y_pred_linear = lin_reg.predict(test[['Area','Room','Parking','Warehouse','Elevator','Address']])
# print mse & r2:
mean_squared_error(test['Price'], y_pred_linear), r2_score(test['Price'], y_pred_linear)


```

```python
# quadratic regression:
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics._regression import mean_squared_error,r2_score
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(train[['Area','Room','Parking','Warehouse','Elevator','Address']])
pol_reg = LinearRegression()
pol_reg.fit(X_poly, train['Price'])
y_pred_poly = pol_reg.predict(poly_reg.fit_transform(test[['Area','Room','Parking','Warehouse','Elevator','Address']]))
# print mse, r2
mean_squared_error(test['Price'], y_pred_poly), r2_score(test['Price'], y_pred_poly)
```

```python
# cubic regression:
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics._regression import mean_squared_error,r2_score
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(train[['Area','Room','Parking','Warehouse','Elevator','Address']])
pol_reg = LinearRegression()
pol_reg.fit(X_poly, train['Price'])
y_pred_poly = pol_reg.predict(poly_reg.fit_transform(test[['Area','Room','Parking','Warehouse','Elevator','Address']]))
# print mse, r2
mean_squared_error(test['Price'], y_pred_poly), r2_score(test['Price'], y_pred_poly)
```

Going from linear to quadratic regression improved the MSE and R2 score. Going from quadratic to cubic regression did not improve the MSE and R2 score as much as going from linear to quadratic regression. This is probably because the data is not cubic in nature. The data is probably quadratic in nature.

```python
# decision tree classifier:
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

# make dtc based on entropy:
dtc = DecisionTreeClassifier(criterion='entropy')
dtc.fit(train[['Area','Room','Parking','Warehouse','Elevator','Address']], train['priceLevel'])
y_pred_dtc = dtc.predict(test[['Area','Room','Parking','Warehouse','Elevator','Address']])
# print accuracy, confusion matrix, classification report:
accuracy_score(test['priceLevel'], y_pred_dtc), confusion_matrix(test['priceLevel'], y_pred_dtc), classification_report(test['priceLevel'], y_pred_dtc, output_dict=True)
```

```python
# random forest classifier using entropy:
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

rfc = RandomForestClassifier(criterion='entropy')
rfc.fit(train[['Area','Room','Parking','Warehouse','Elevator','Address']], train['priceLevel'])
y_pred_rfc = rfc.predict(test[['Area','Room','Parking','Warehouse','Elevator','Address']])
# print accuracy, confusion matrix, classification report:
accuracy_score(test['priceLevel'], y_pred_rfc), confusion_matrix(test['priceLevel'], y_pred_rfc), classification_report(test['priceLevel'], y_pred_rfc, output_dict=True)
```

```python
# knn classifier, k = 5:
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(train[['Area','Room','Parking','Warehouse','Elevator','Address']], train['priceLevel'])
y_pred_knn = knn.predict(test[['Area','Room','Parking','Warehouse','Elevator','Address']])
# print accuracy, confusion matrix, classification report:
accuracy_score(test['priceLevel'], y_pred_knn), confusion_matrix(test['priceLevel'], y_pred_knn), classification_report(test['priceLevel'], y_pred_knn, output_dict=True)
```

```python
# knn classifier, k = 10:
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(train[['Area','Room','Parking','Warehouse','Elevator','Address']], train['priceLevel'])
y_pred_knn = knn.predict(test[['Area','Room','Parking','Warehouse','Elevator','Address']])
# print accuracy, confusion matrix, classification report:
accuracy_score(test['priceLevel'], y_pred_knn), confusion_matrix(test['priceLevel'], y_pred_knn), classification_report(test['priceLevel'], y_pred_knn, output_dict=True)
```

```python
# knn classifier, k = 3:
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train[['Area','Room','Parking','Warehouse','Elevator','Address']], train['priceLevel'])
y_pred_knn = knn.predict(test[['Area','Room','Parking','Warehouse','Elevator','Address']])
# print accuracy, confusion matrix, classification report:
accuracy_score(test['priceLevel'], y_pred_knn), confusion_matrix(test['priceLevel'], y_pred_knn), classification_report(test['priceLevel'], y_pred_knn, output_dict=True)
```

```python
# SVM classifier:
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

svm = SVC()
svm.fit(train[['Area','Room','Parking','Warehouse','Elevator','Address']], train['priceLevel'])
y_pred_svm = svm.predict(test[['Area','Room','Parking','Warehouse','Elevator','Address']])
# print accuracy, confusion matrix, classification report:
accuracy_score(test['priceLevel'], y_pred_svm), confusion_matrix(test['priceLevel'], y_pred_svm), classification_report(test['priceLevel'], y_pred_svm, output_dict=True)
```

From these, based on the accuracy alone, decision tree seems best for this dataset, and it also has the best precision/recall metrics in the dataset.

```python
# neural network classifier using tensorflow:
import tensorflow as tf


# define the model:
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(12, activation='relu'),
    tf.keras.layers.Dense(12, activation='relu'),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax')
])

# compile the model:
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
# make one-hot encoding for the labels using tensorflow:
train_labels = tf.keras.utils.to_categorical(train['priceLevel'])
test_labels = tf.keras.utils.to_categorical(test['priceLevel'])
# train the model:
history = model.fit(train[['Area','Room','Parking','Warehouse','Elevator','Address']], train_labels, epochs=30)
# predict:
y_pred_nn = model.predict(test[['Area','Room','Parking','Warehouse','Elevator','Address']])
# print accuracy and confusion_metrics
tf.math.confusion_matrix(test_labels.argmax(axis=1), y_pred_nn.argmax(axis=1)), accuracy_score(test_labels.argmax(axis=1), y_pred_nn.argmax(axis=1))

```

```python
# plot model history
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['loss'])
plt.title('model accuracy and loss')
plt.ylabel('accuracy/loss')
plt.xlabel('epoch')
plt.legend(['accuracy', 'loss'], loc='upper left')
plt.show()
```
